```python
import numpy as np  # For numerical operations
import pandas as pd  # For handling datasets
from sklearn.model_selection import train_test_split  # For splitting data into train and test sets
from sklearn.linear_model import LogisticRegression  # For Logistic Regression model
from sklearn.ensemble import RandomForestClassifier  # For Random Forest model
import xgboost as xgb  # For XGBoost model
from sklearn.metrics import accuracy_score, classification_report  # For model evaluation
from sklearn.ensemble import VotingClassifier  # For creating an ensemble model
```

```python
# Load  dataset, assuming 'data' is your dataset with features and labels
data = pd.read_csv(r'D:\data analytics\Case study 3 - papaya\DAY#3 - Unsupervised ML & Use Cases\Use Case Dataset\CUSTOMER ACQUISITION PREDICTION\Use Case #1.csv')

```

```python
data
```

```python
features = data[['store_sales(in millions)', 'store_cost(in millions)', 'unit_sales(in millions)']]
labels = data['marital_status']
labels = labels.map({'M': 0, 'S': 1})

```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
# Step 1: Create an instance of Logistic Regression model
logistic_reg = LogisticRegression(random_state=42)
```

```python
# Step 2: Fit the Logistic Regression model on the training data
logistic_reg.fit(X_train, y_train)
```

```python
# Step 3: Create an instance of Random Forest model
random_forest = RandomForestClassifier(random_state=42)
```

```python
# Step 4: Fit the Random Forest model on the training data
random_forest.fit(X_train, y_train)
```

```python
# Step 5: Create an instance of XGBoost model
xgb_model = xgb.XGBClassifier(random_state=42)
```

```python
# Step 6: Fit the XGBoost model on the training data
xgb_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set for each model

# Step 7: Make predictions using the Logistic Regression model
logistic_reg_preds = logistic_reg.predict(X_test)
```

```python
# Step 8: Make predictions using the Random Forest model
random_forest_preds = random_forest.predict(X_test)
```

```python
# Step 9: Make predictions using the XGBoost model
xgb_preds = xgb_model.predict(X_test)
```

```python
# Step 10: Combine models using Voting Classifier
ensemble_model = VotingClassifier(estimators=[
    ('logistic_regression', logistic_reg),
    ('random_forest', random_forest),
    ('xgboost', xgb_model)],
    voting='hard')  # Use 'hard' voting for majority rule
```

```python
# Step 11: Fit the ensemble model on the training data
ensemble_model.fit(X_train, y_train)

```

```python
# Make predictions with the ensemble model
ensemble_preds = ensemble_model.predict(X_test)
```

```python
# Step 12: Evaluate each model and the ensemble

# Evaluate Logistic Regression model
logistic_reg_accuracy = accuracy_score(y_test, logistic_reg_preds)
logistic_reg_classification_report = classification_report(y_test, logistic_reg_preds)
```

```python
# Evaluate Random Forest model
random_forest_accuracy = accuracy_score(y_test, random_forest_preds)
random_forest_classification_report = classification_report(y_test, random_forest_preds)
```

```python
# Evaluate XGBoost model
xgb_accuracy = accuracy_score(y_test, xgb_preds)
xgb_classification_report = classification_report(y_test, xgb_preds)
```

```python
# Evaluate Ensemble model
ensemble_accuracy = accuracy_score(y_test, ensemble_preds)
ensemble_classification_report = classification_report(y_test, ensemble_preds)
```

```python
# Print the evaluation results for each model and the ensemble
print("Logistic Regression Accuracy:", logistic_reg_accuracy)
print("Logistic Regression Classification Report:\n", logistic_reg_classification_report)
print("Random Forest Accuracy:", random_forest_accuracy)
print("Random Forest Classification Report:\n", random_forest_classification_report)
print("XGBoost Accuracy:", xgb_accuracy)
print("XGBoost Classification Report:\n", xgb_classification_report)
print("Ensemble Model Accuracy:", ensemble_accuracy)
print("Ensemble Model Classification Report:\n", ensemble_classification_report)
```
